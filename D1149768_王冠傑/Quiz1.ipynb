```python
# quiz4:

# 給定字串如下，基於自定義置左及置中，以及填補空位的字元
# 參考https://www.runoob.com/python/att-string-format.html
# 分別印出以下結果:
#  1. 佔 15 格置左，空閒格子補 _         
#  2. 佔 15 格置中
#  3. 佔 15 格置中，空閒格子補 *
#  4. 佔 15 格置右
#  5. 佔 15 格置右，空閒格子補 ~

S = "abcde"
print(f'{S:_<15s}')
print(f'{S:^15s}')
print(f'{S:*^15s}')
print(f'{S:>15s}')
print(f'{S}')
      
```
